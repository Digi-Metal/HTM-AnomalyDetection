```python
import numpy as np
import time
import csv
import glob
import datetime
```

```python
def data_generator():
    x1 = np.random.uniform(0, 15)
    x2 = np.random.uniform(0, 15)
    return x1, x2

# def count_rows(filepath):
#     count = -1
#     for count, line in enumerate(open(filepath, 'r')):
#         pass
#     count += 1
#     return count

# def write_to_file(data, filepath):
#     f = open(filepath, "a")
#     f.write(data)
#     f.close()


def count_rows(filepath):
    with open(filepath, 'r') as f:
        return len(f.readlines())


def clear_top_1_data(filepath):
    csvfile = glob.glob(filepath)
    for f in csvfile: 
        lines = open(f).readlines() 
        open(f, 'wb').writelines(lines[0:2])
        open(f, 'wb').writelines(lines[4:])


def write_to_csv(data, filepath):
    '''
    This function has been defined for csv file writing.
        --outputfile: The output csv file
        --row: The rows write to csv file
    
    '''
    csvfile = open(filepath, 'wb+')
    writer = csv.writer(csvfile)
    writer.writerows(data)
    csvfile.close()    

    
def time_arithmetic(date):
    delta = datetime.timedelta(hours=1)
    out_date = date + delta
    return out_date
```
