```python
import xlrd
import csv
import time
from datetime import datetime
```

```python
file = 'D:\\chemical_data.xlsx'

def read_excel(file):
    wb = xlrd.open_workbook(filename=file) # open this file

#     print wb.sheet_names() # print the name of all sheets

    sheet1 = wb.sheet_by_name('Feed Rate')
    sheet2 = wb.sheet_by_name('Feed Blend') # find the sheet via its name
    sheet3 = wb.sheet_by_name('Moisture')

    # deal with sheet1 - Feed rate
    sheet1_nrows = sheet1.nrows
    sheet1_rows = []
    header = sheet1.row_values(2)
    sheet1_rows.append(header)
    for i in range(4, sheet1_nrows):      
        row = sheet1.row_values(i)
        sheet1_rows.append(row)
        
    # deal with sheet2 - Feed Blend
    sheet2_nrows = sheet2.nrows
    sheet2_rows = []
    header = sheet2.row_values(2)
    sheet2_rows.append(header)
    for i in range(4, sheet2_nrows):      
        row = sheet2.row_values(i)
        sheet2_rows.append(row)    
    
    # deal with sheet3 - Moisture
    sheet3_nrows = sheet3.nrows
    sheet3_rows = []
    header = sheet3.row_values(2)
    sheet3_rows.append(header)
    for i in range(4, sheet3_nrows):      
        row = sheet3.row_values(i)
        sheet3_rows.append(row)
        
    return sheet1_rows, sheet2_rows, sheet3_rows


def timestamp_transfer(timestamp):
    default_date = '2019-8-7 '
    date_string = default_date + timestamp
    date_time = datetime.strptime(date_string, '%Y-%m-%d %H:%M:%S')
    return date_time.strftime('%Y-%m-%d %H:%M:%S')

def input_output_spliter(sheet_rows):
    input_variables = []
    output_variables = []
    
    output_header = sheet_rows[0][2:9]
    output_header.append('Time') # add timestamp
    output_header.append(sheet_rows[0][13]) # add label
    input_header = sheet_rows[0][9:13]
    input_header.append('Time')
    input_header.append(sheet_rows[0][13])
    
    output_data_types = ['float', 'float', 'float', 'float', 'float', 'float', 'float', 'datetime', 'int']
    output_tag = ['', '', '', '', '', '', '', 'T', '']
    input_data_types = ['float', 'float', 'float', 'float', 'datetime', 'int']
    iutput_tag = ['', '', '', '', 'T', '']
    
    output_variables.append(output_header)
    input_variables.append(input_header)
    output_variables.append(output_data_types)
    output_variables.append(output_tag)
    input_variables.append(input_data_types)
    input_variables.append(iutput_tag)
    
    for i in range(1, len(sheet_rows)):
        row = sheet_rows[i]
        outputv = row[2:9]
        inputv = row[9:13]
        timestamp = row[0]
        timestamp = timestamp_transfer(timestamp)
        label = [int(row[13])]
        outputv.append(timestamp)
        outputv.extend(label)
        inputv.append(timestamp)
        inputv.extend(label)     
        output_variables.append(outputv)
        input_variables.append(inputv)
    return input_variables, output_variables


def CSVWriter(outputfile, row):
    csvfile = open(outputfile, 'a+')
    writer = csv.writer(csvfile)
    writer.writerows(row)
    csvfile.close()
```

```python
input_feed_rate = 'D:\\chemical_data\\input_feed_rate.csv'
output_feed_rate = 'D:\\chemical_data\\output_feed_rate.csv'
input_feed_blend = 'D:\\chemical_data\\input_feed_blend.csv'
output_feed_blend = 'D:\\chemical_data\\output_feed_blend.csv'
input_moisture = 'D:\\chemical_data\\input_moisture.csv'
output_moisture = 'D:\\chemical_data\\output_moisture.csv'

sheet1_rows, sheet2_rows, sheet3_rows = read_excel(file)

# saving sheet1 to csv
inn, outt = input_output_spliter(sheet1_rows)
CSVWriter(input_feed_rate, inn)
CSVWriter(output_feed_rate, outt)

# saving sheet2 to csv
inn, outt = input_output_spliter(sheet2_rows)
CSVWriter(input_feed_blend, inn)
CSVWriter(output_feed_blend, outt)

# saving sheet3 to csv
inn, outt = input_output_spliter(sheet3_rows)
CSVWriter(input_moisture, inn)
CSVWriter(output_moisture, outt)
```
