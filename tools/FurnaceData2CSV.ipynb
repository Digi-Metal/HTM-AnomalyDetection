```python
import xlrd
import csv
import time
from datetime import datetime
```

```python
file = 'D:\\chemical_data.xlsx'

def read_excel(file):
    wb = xlrd.open_workbook(filename=file) # open this file

#     print wb.sheet_names() # print the name of all sheets

    sheet1 = wb.sheet_by_name('Feed Rate')
    sheet2 = wb.sheet_by_name('Feed Blend') # find the sheet via its name
    sheet3 = wb.sheet_by_name('Moisture')

    # deal with sheet1 - Feed rate
    sheet1_nrows = sheet1.nrows
    sheet1_rows = []
    header = sheet1.row_values(2)
    sheet1_rows.append(header)
    for i in range(4, sheet1_nrows):      
        row = sheet1.row_values(i)
        sheet1_rows.append(row)
        
    # deal with sheet2 - Feed Blend
    sheet2_nrows = sheet2.nrows
    sheet2_rows = []
    header = sheet2.row_values(2)
    sheet2_rows.append(header)
    for i in range(4, sheet2_nrows):      
        row = sheet2.row_values(i)
        sheet2_rows.append(row)    
    
    # deal with sheet3 - Moisture
    sheet3_nrows = sheet3.nrows
    sheet3_rows = []
    header = sheet3.row_values(2)
    sheet3_rows.append(header)
    for i in range(4, sheet3_nrows):      
        row = sheet3.row_values(i)
        sheet3_rows.append(row)
        
    return sheet1_rows, sheet2_rows, sheet3_rows


def input_output_spliter(sheet_rows):
    input_variables = []
    output_variables = []
    for row in sheet_rows:
        timestamp = row[0]
        outputv = row[2:9]
        inputv = row[9:14]
        outputv.extend(timestamp)
        inputv.extend(timestamp)
        output_variables.append(outputv)
        input_variables.append(inputv)
    return input_variables, output_variables


def CSVWriter(outputfile, row):
    csvfile = open(outputfile, 'a+')
    writer = csv.writer(csvfile)
    writer.writerows(row)
    csvfile.close()
```

```python
sheet1_rows, sheet2_rows, sheet3_rows = read_excel(file)
print len(sheet1_rows), len(sheet2_rows), len(sheet3_rows)
```
